PART 1

```python
import folium
import requests
import pandas

arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")

arrest_table
```

```python
arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
```

PART 2

```python
map_osm = folium.Map(location = [39.29, -76.61], zoom_start = 13)
map_osm
```

PART 3

```python
def setColor(s):
    if s == 'M':
        fill_color = 'blue'
    else:
        fill_color = 'pink'
    return fill_color

def setHtml(row): 
    html = "Arrest #: " + str(row['arrest']) + "\n" + \
    "Race: " + row['race'] + "\n" + \
    "Age: " + str(row['age']) + "\n" + \
    "Incident Offense: " + row['incidentOffense']
    return html

for col, row in arrest_table.sample(250).iterrows():

    folium.Marker(location = [row['lat'], row['long']], popup = setHtml(row),
                      icon = folium.Icon(color = setColor(row['sex']))).add_to(map_osm)

map_osm
```

This is an interactive criminal data map in baltimore. I pick up 250 samples and map the sample using markers that indicate the sex of the offender. I use blue for male and pink for female, and we can see that the difference of gender of offender, which is useful for criminology, sociobiology, and even feminist studies.
